```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np
%matplotlib inline
plt.rcParams["figure.dpi"]=150
plt.rcParams["figure.facecolor"]="white"
plt.rcParams["figure.figsize"]=(4,3)
```

```python
bohr2ang = 0.52917721
x, y = np.loadtxt('scan.csv', delimiter=',', unpack=True)
x=x*bohr2ang
plt.plot(x, y, "o-", markersize=5, label='Potential energy curve')
plt.xlabel('a (Angstrom)')
plt.ylabel('Total Energy (Ry)')
plt.xlim([3.0,4.0])
plt.ylim([-36.9,-36.5])
plt.grid()
plt.legend(frameon=False)
plt.show()
```

The equilibrium lattice constant seems to be around 3.6 Angstrom, which is very close to the experimental value of 3.57 Angstrom as seen from this reference [https://iopscience.iop.org/article/10.7567/JJAP.57.111301](https://iopscience.iop.org/article/10.7567/JJAP.57.111301)
